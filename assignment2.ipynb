```python
# Import necessary libraries
import vertexai
from vertexai.language_models import TextGenerationModel
from IPython.display import display, clear_output
import ipywidgets as widgets


# Initialize Vertex AI with your project and location
vertexai.init(project="assinment-2-1", location="us-central1")
model = TextGenerationModel.from_pretrained("text-bison-32k")

def summarise(text):
  parameters = {
        "candidate_count": 1,
        "max_output_tokens": 2048,
        "temperature": 0.9,
        "top_p": 1
    }
  prompt = f"""Topic: Cloud Computing
      Contentsourse: {text}
      summarise all the contents so i can use it to create quiz later
  """
  response = model.predict(prompt, **parameters)
  return response.text

# Define the function to generate the quiz using your model
def generate_quiz( number_of_questions,quiz_type_input, difficulty_level, language):
    parameters = {
        "candidate_count": 1,
        "max_output_tokens": 8192,
        "temperature": 0.9,
        "top_p": 1
    }
    prompt = f"""Topic: Cloud Computing
    Contentsourse: {summarised_contents}
    Number of questions: {number_of_questions}
    Difficulty level: {difficulty_level}
    Language: {language}
    Mixture of Quiz types: {quiz_type_input}
    show all answer at once after the last quesitons
"""

    response = model.predict(prompt, **parameters)
    print(f"Response from Model: {response.text}")

# Create widgets for inputs
number_of_questions_input = widgets.Dropdown(options=["5", "10", "15","20"], value="10", description="Number of Questions:")
quiz_type_input = widgets.SelectMultiple(options=["multiple choice", "short answer","true false"], value=["multiple choice"], description="Quiz Type:")
difficulty_level_input = widgets.Dropdown(options=["easy", "intermediate", "hard"], value="intermediate", description="Difficulty Level:")
language_input = widgets.Dropdown(options=["English", "Spanish", "French"], value="English", description="Language:")
generate_button = widgets.Button(description="Generate Quiz")

# Function to handle button click
def on_generate_button_clicked(b):
    # Clear previous outputs using the correct function from IPython.display
    clear_output()
    # Generate the quiz with the provided inputs
    generate_quiz( number_of_questions_input.value,quiz_type_input.value, difficulty_level_input.value, language_input.value)


# Add a callback to the button
summarised_contents =[]
for i in range(len(file_contents_array)):
  summarised_contents.append(summarise(file_contents_array[i]))

generate_button.on_click(on_generate_button_clicked)

# Display the widgets
display( number_of_questions_input,quiz_type_input, difficulty_level_input, language_input, generate_button)
```
